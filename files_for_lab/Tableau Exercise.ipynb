```python
import numpy as np
import pandas as pd

def es_discreta(data):
   
    if np.issubdtype(data.dtype, np.integer):
        return True
    
    elif np.issubdtype(data.dtype, np.bool_):
        return True
  
    elif np.issubdtype(data.dtype, np.str_) or np.issubdtype(data.dtype, np.unicode_):
        return True
    
    else:
        return False

for columna in data.columns:
    columna_data = data[columna]
    es_columna_discreta = es_discreta(columna_data)
    
    if es_columna_discreta:
        print(f" '{columna}'has discret data.")
    else:
        print(f"'{columna}' has continuos data.")
```
